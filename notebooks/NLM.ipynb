```python
import sys
sys.path.append('../src')
import cv2
import NLM
from matplotlib import pyplot as plt

# Cargar imagen
img = cv2.imread('nlm1.png')

# Aplicar filtro NLM
filtered = NLM.filter(img, h=10)

# Mostrar comparacin
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original')
axes[0].axis('off')
axes[1].imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
axes[1].set_title('NLM Filtered')
axes[1].axis('off')
plt.show()
```
